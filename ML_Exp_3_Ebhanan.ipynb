```python
# @title Salary vs. Years of Experience by Salary Category

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsExperience', y='Salary', hue='Salary_Category', data=df, alpha=0.7)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
_ = plt.title('Salary vs. Years of Experience by Salary Category')
```

```python
# @title Proportion of Employees in Each Salary Category

import matplotlib.pyplot as plt

category_counts = df['Salary_Category'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
_ = plt.title('Proportion of Employees in Each Salary Category')
```

```python
# @title Distribution of Years of Experience

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['YearsExperience'], bins=10, edgecolor='black')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
_ = plt.title('Distribution of Years of Experience')
```

```python
# @title Salary Distribution by Experience Level

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.boxplot(x='Salary_Category', y='Salary', data=df)
plt.xlabel('Salary Category')
plt.ylabel('Salary')
_ = plt.title('Salary Distribution by Experience Level')
```

```python
# @title YearsExperience

from matplotlib import pyplot as plt
df['YearsExperience'].plot(kind='hist', bins=20, title='YearsExperience')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Treemap (requires plotly)
fig = px.treemap(df, path=['YearsExperience'], values='Salary',title='Salary distribution across years of experience')
fig.show()
```

```python
# Bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['YearsExperience'], df['Salary'], s=df['YearsExperience'] * 10, alpha=0.5)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Bubble Chart of Salary vs. Years of Experience')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Salary'], bins=10, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Salary')
plt.show()
```

```python
# Column chart (bar chart)
plt.figure(figsize=(8, 6))
plt.bar(df['YearsExperience'], df['Salary'])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title('Salary vs. Years of Experience')
plt.show()
```

```python
# Line chart
plt.figure(figsize=(8, 6))
plt.plot(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience (Scatterplot)')
plt.show()
```
